```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
movie = movie.drop(['wiki_link','title_y','original_title','is_adult','year_of_release','runtime','imdb_rating','imdb_votes','release_date'], axis=1)
```

```python
movie = movie.drop(['tagline','wins_nominations'],axis=1)
```

```python
movie = movie.drop(['story'], axis=1)
```

```python
movie
```

```python
movie.isnull().sum()
```

```python
movie = movie.dropna()
```

```python
movie.head()
```

```python
movie.tail()
```

```python
movie = movie.reset_index(drop=True)
```

```python
movie.tail()
```

```python
movie.head()
```

```python
movie.shape
```

```python
movie.isnull().sum()
```

```python
movie['genres'] = movie['genres'].str.split('|')
```

```python
movie['actors'] = movie['actors'].str.split('|')
```

```python
movie['actors'][0]
```

```python
for i in range(len(movie['actors'])):
    if len(movie['actors'][i]) > 4:    
        movie['actors'][i] = movie['actors'][i][0:4]
```

```python
movie.head()
```

```python
def merg(text):
    merge = []
    for i in text:
        merge.append(i.replace(" ",""))
    return merge
```

```python
movie['actors'] = movie['actors'].apply(merg)
```

```python
movie['summary'] = movie['summary'].str.split()
```

```python
movie.head()
```

```python
movie['tags'] = movie['genres'] + movie['summary'] + movie['actors']
```

```python
movie.head()
```

```python
movie['tags'] = movie['tags'].apply(lambda x: ' '.join(x))
```

```python
movie['tags'][0]
```

```python
movie.head()
```

```python
movie['title'] = movie['title_x'].rename("title")
```

```python
movie.head()
```

```python
df = movie[['imdb_id', 'title', 'poster_path', 'tags']]
```

```python
df.head()
```

```python
df['tags'] = df['tags'].apply(lambda x: x.lower())
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=4000, stop_words='english')
```

```python
vector = cv.fit_transform(df['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```
